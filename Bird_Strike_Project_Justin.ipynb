```python
# We'll first import pandas
# it is standard to import it as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
#CSV Bird Strikes from 2000-2019
df = pd.read_csv("wildlife.csv")
```

```python
#Displaying All columns and trying to display rows as well
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 1000)
df.head()
```

```python
#Copying the dataset
df_copy=df.copy()
```

```python
#Dropping columns that I percieved as not worth looking into 
df_copy.drop(columns=["TYPE_ENG","AMA","AMO","EMA","EMO","AC_CLASS","AC_MASS","REG","REMAINS_COLLECTED","REMAINS_SENT","RUNWAY","STR_RAD","DAM_RAD","STR_WINDSHLD","STR_NOSE","DAM_NOSE","STR_ENG1","DAM_ENG1","STR_ENG2","DAM_ENG2","STR_ENG3","DAM_ENG3","STR_ENG4","DAM_ENG4","STR_PROP","DAM_PROP","STR_WING_ROT","DAM_WING_ROT","STR_FUSE","DAM_FUSE","STR_LG","DAM_LG","STR_TAIL","DAM_TAIL","STR_LGHTS","DAM_LGHTS","STR_OTHER","DAM_OTHER","OTHER_SPECIFY","REPORTED_NAME","REPORTED_TITLE","TRANSFER", "Unnamed: 101"])
```

```python
#Total cost direct and indirect as the result of a bird strike
total = df_copy['COST_REPAIRS'].fillna(0)+df_copy['OTHER_COST'].fillna(0)
df_copy['TOTAL']=total
```

```python
#From 2000-2019 the total in damages 
df_copy['TOTAL'].sum()
```

```python
#Number of species hit
df_copy.SPECIES.value_counts()
```

```python
#Number of collisions per year
df_count = df_copy['INCIDENT_YEAR'].value_counts(sort=False).rename_axis('Year').reset_index(name='counts')
print(df_count)
```

```python
#Plot of collisions per year, my guess is that the trend is most likely due to an increase in Air travel
plt.figure(figsize=(10,10))

plt.plot(df_count.Year,df_count.counts)
plt.show()
```

```python
#Total number of collisions per airline from 2000-2019
df_copy.OPERATOR.value_counts()
```

```python
#Total number of collisions per airport from 2000-2019
df_count1 = df_copy['AIRPORT'].value_counts(sort=True).rename_axis('Date').reset_index(name='counts')
print(df_count1)
```

```python
#Top 100 airports
print(df_count1.Date[0:100])
```

```python
print(df_copy.columns)

#Changing the formating of the date entries
df_copy['month'] = pd.to_datetime(df_copy['INCIDENT_DATE '])
df_copy.head()
```

```python
#Grouping by date
a=df_copy.groupby('month').count()
b=a['INDX_NR']
print(b)
```

```python
#Adding an index 
b1 = b.rename_axis('Date').reset_index(name='counts')
print (b1)
```

```python
#Number of collisions per day from 2000-2019
plt.figure(figsize=(10,10))

plt.plot(b1.Date,b1.counts)
#plt.xlim(b1[[0]],b1[[2000]])
plt.xlim(pd.Timestamp('2000-01-01'), pd.Timestamp('2020-01-01'))
plt.show()
```

```python
#Grouping the total cost in damages by day
a=df_copy.groupby('month').sum()
b=a['TOTAL']
print(b)
```

```python
#Adding an index
b2 = b.rename_axis('Date').reset_index(name='Total')
print (b2)
```

```python
#Plotting the total cost 
plt.figure(figsize=(10,10))

plt.plot(b2.Date,b2.Total)
#plt.xlim(b1[[0]],b1[[2000]])
plt.ylim(0,10000000)
plt.xlim(pd.Timestamp('2000-01-01'), pd.Timestamp('2020-01-01'))
plt.show()
```
