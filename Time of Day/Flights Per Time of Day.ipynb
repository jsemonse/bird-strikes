## Flights Per Time of Day

In this notebook, we examine the number of flights per time of day. That is, we plot the number of flights scheduled to depart within each hour of the day (e.g. flights departing between 3:00 AM and 3:59 AM). This data comes from the FAA Aviation System Performance Metrics site: https://aspm.faa.gov/apm/sys/AnalysisAP.asp

The main purpose of this notebook is to become familiar with the hourly flight data and how it is formatted. As a toy example, we work with data from 2018. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# The first line of the data file from FAA ASPM contains title information, e.g.
# "ASPM : Airport Analysis : All Flights ReportFrom 01/2018 To 12/2018 : Use Flight Plan"
# The last line of the data file is a "Total:" row
# Use skiprows = [0, 26] to avoid reading in these two lines
flights_2018 = pd.read_csv("flights_hourly/flights_hourly_2018.csv", 
                           skiprows = [0, 26])
```

```python
plt.figure(figsize = (10,10))
plt.bar(flights_2018['Hour'], flights_2018['ScheduledDepartures'])
plt.xlabel('Hour', fontsize = 14)
plt.ylabel('Number of flights (2018)', fontsize = 14)
plt.xticks(flights_2018['Hour'])
plt.show()
```
