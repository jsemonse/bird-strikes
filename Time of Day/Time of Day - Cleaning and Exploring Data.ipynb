## Cleaning and Exploring the Time-of-Day Data

This notebook is for playing around with the 'TIME' column of the bird strike data found in the "strikes_2000_2019.csv" file. The 'TIME' column records the time of day that the strike occured.

```python
import numpy as np
import pandas as pd
import pprint
```

```python
# Load in the strike data
strikes = pd.read_csv("strikes_2000_2019.csv")
```

```python
# Print all the column names
print([strikes.columns[i] for i in range(len(strikes.columns))])
```

```python
# Make a smaller data frame to focus on the strike time and place
strike_times = strikes[['INDX_NR','INCIDENT_DATE ','STATE',
                        'AIRPORT_ID', 'OPERATOR_ID','SPECIES_ID',
                        'DAMAGE','COST_REPAIRS',
                        'INCIDENT_MONTH','INCIDENT_YEAR',
                        'TIME_OF_DAY','TIME',
                        'FAAREGION','ENROUTE','LOCATION',
                        'HEIGHT','SPEED','DISTANCE',
                        'PHASE_OF_FLIGHT',
                        'EFFECT_NONE', 'EFFECT_ABORTED', 
                        'EFFECT_PREC_LAND', 'EFFECT_ENG_SHUT', 
                        'EFFECT_OTH', 'EFFECT_OTHER',
                        'SKY', 'PRECIP_FOG', 'PRECIP_RAIN', 
                        'PRECIP_SNOW', 'PRECIP_NONE', 
                        'BIRDS_SEEN', 'BIRDS_STRUCK', 'SIZE', 
                        'WARNED', 'REMARKS', 'AOS',
                        'OTHER_COST', 'COST_REPAIRS_INFL_ADJ', 
                        'COST_OTHER_INFL_ADJ','COMMENT',
                        'INDICATED_DAMAGE']]
```

```python
# print an example row
# note that not all of them have a time of day
strike_times.iloc[11,:]
```

```python
print(strike_times['TIME_OF_DAY'].value_counts(dropna=False))
```

```python
strike_times['TIME'].value_counts()
# note that whitespace appears 92325 times,
# more than any other value
```

```python
# print out the blank value at the top of the list above
strike_times['TIME'].value_counts().index[0]
```

```python
# apparently blank values in the TIME column consist of 5 spaces
len(strike_times.loc[strike_times['TIME']=='     '])
for i in range(0,20):
    print(i, len(strike_times.loc[strike_times['TIME']==i*' ']))
```

```python
# Replace strings of 5 spaces with NaN
strike_times = strike_times.replace('     ', np.nan)
```
